```python
import json

import numpy as np
import keras
from matplotlib import pyplot as plt
```

```python
with open('../models/history-synthetic-data.json') as f:
    history_syn = json.load(f)
```

```python
width = 0.3 
plt.figure(figsize=(15, 5))
plt.bar(np.arange(30) + 1 - width / 2, history_syn['loss'], width, alpha=0.5, color='C2')
plt.bar(np.arange(30) + 1 + width / 2, history_syn['val_loss'], width, alpha=0.5, color='C3')
plt.title('Значення функції втрат OCR моделі', fontsize=16)
plt.xlabel('номер епохи тренування', fontsize=16)
plt.legend(['синтетичні тренувальні дані', 'синтетичні валідаційні дані'], loc='upper right', fontsize=14)
plt.ylim([0., 15.])
plt.grid()
plt.savefig('syn_loss.png')
```

```python
history_orig = {'loss': [], 'val_loss': []}

with open('../models/history-original.txt', 'r') as f:
    for l in f.readlines():
        if 'loss' in l:
            loss = float(l[l.find(' loss:') + 7: l.find(' loss:') + 14])
            val_loss = float(l[l.find('_loss:') + 7: l.find('_loss:') + 14])
            history_orig['loss'].append(loss)
            history_orig['val_loss'].append(val_loss)
```

```python
width = 0.3 
plt.figure(figsize=(15, 5))
plt.bar(np.arange(100) + 1 - width / 2, history_orig['loss'], width, alpha=0.5)
plt.bar(np.arange(100) + 1 + width / 2, history_orig['val_loss'], width, alpha=0.5)
plt.title('Значення функції втрат OCR моделі', fontsize=16)
plt.xlabel('номер епохи тренування', fontsize=16)
plt.legend(['оригінальні тренувальні дані', 'оригінальні валідаційні дані'], loc='upper right', fontsize=14)
plt.ylim([0., 15.])
plt.grid()
plt.savefig('orig_loss.png')
```
