```python
from sklearn import preprocessing
from sklearn import svm
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
import librosa
import numpy as np
import os


```

```python
folder_path = "/content/drive/MyDrive/AI/Audio/Train"
mfccs = []
languages = []
```

```python
for filename in os.listdir(folder_path):
    # завантажуємо аудіофайл
    audio_data, sampling_rate = librosa.load(os.path.join(folder_path, filename), sr=None)
    
    # обчислюємо MFCC для цього аудіофайлу
    mfcc = librosa.feature.mfcc(y=audio_data, sr=sampling_rate)
    
    # беремо середнє значення по кожному коефіцієнту MFCC
    mfcc_mean = np.mean(mfcc, axis=1)
    
    # додаємо це до нашого списку mfcc
    mfccs.append(mfcc_mean)
    
    # додаємо відповідну мову до списку мов
    # це припущення, що назва мови міститься в назві файла
    # ви маєте змінити це відповідно до вашого випадку
    language = filename.split("_")[0]
    languages.append(language)
```

```python
# Розділіть дані на навчальний та тестовий набори
X_train, X_test, y_train, y_test = train_test_split(mfccs, languages, test_size=0.2, random_state=42)
```

```python
X_train, X_test, y_train, y_test
```

```python
# Створіть SVM класифікатор
clf = svm.SVC(kernel='linear') 
```

```python
# Навчіть модель
clf.fit(X_train, y_train)
```

```python
# Протестуйте модель на тестовому наборі
y_pred = clf.predict(X_test)

for pred, test in zip(y_pred, y_test):
    print(f'Result : {pred == test},\tPredicted:\t{pred},\tActual: {test} ')
```

```python
# Виведіть результати
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
